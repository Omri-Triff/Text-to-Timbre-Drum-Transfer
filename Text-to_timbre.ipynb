```python
# Patch app.py to enable public sharing link for Gradio
import os

with open("app.py", "r") as f:
    content = f.read()

# מחליף את הפקודה launch הרגילה בגרסה שיוצרת לינק ציבורי
if "launch(share=True)" not in content:
    content = content.replace(".launch()", ".launch(share=True)")
    content = content.replace("launch(server_name", "launch(share=True, server_name") # ליתר ביטחון

with open("app.py", "w") as f:
    f.write(content)

print("App patched for Colab!")
```
